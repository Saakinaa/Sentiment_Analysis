# **Social Media : Twitter API and Sentiment Score**

```python
import tweepy
import pandas as pd
import numpy as np

# Clés APIs
bearer_Token = "bearer_token"
consumer_key = "consumer_key"
consumer_secret = "consumer_secret"
access_token = "access_token"
access_token_secret = "access_token_secret"

import tweepy

# Authentification avec tweepy Client (API V2)
client = tweepy.Client(bearer_token=bearer_Token)

# Recherche des tweets sur "senegal"
tweets = client.search_recent_tweets(query="senegal", tweet_fields=['context_annotations', 'created_at'], max_results=100)

# Affichage des tweets
for tweet in tweets.data:
    print(tweet.text)
```

```python
import tweepy
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch

# Load tokenizer and pre-trained model for Sentiment Analysis
tokenizer = AutoTokenizer.from_pretrained("nlptown/bert-base-multilingual-uncased-sentiment")
model = AutoModelForSequenceClassification.from_pretrained("nlptown/bert-base-multilingual-uncased-sentiment")

# Sentiment Analysis Function
def analyze_sentiment(text):
    tokens = tokenizer.encode(text, return_tensors='pt')
    result = model(tokens)
    sentiment = torch.argmax(result.logits).item()
    return sentiment  # Renvoie un score de sentiment entre 0 (négatif) et 4 (positif)

# Sentiments Analysis
for tweet in tweets.data:
    tweet_text = tweet.text
    sentiment_score = analyze_sentiment(tweet_text)
    print(f"Tweet: {tweet_text}")
    print(f"Sentiment Score: {sentiment_score}")
    print("-" * 50)
```
